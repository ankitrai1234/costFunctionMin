```python
import pandas as pd

# Assuming 'df' is your existing DataFrame

# Define columns to exclude from conversion
exclude_columns = ['Item', 'Item.1', 'Grade', 'Date']

# Convert all columns except the excluded ones to float
df.iloc[:, ~df.columns.isin(exclude_columns)] = df.iloc[:, ~df.columns.isin(exclude_columns)].apply(pd.to_numeric, errors='coerce')

# Print the modified DataFrame
df
```

```python
import pandas as pd

# Assuming 'df' is your existing DataFrame
# Replace the list ['Mechanical', 'Mechanical.1', 'Mechanical.2', 'Mechanical.3', 'Microstructure'] 
# with the actual column names you want to interpolate

# Convert specified columns to numeric if they are not already
columns_to_interpolate = ['Mechanical', 'Mechanical.1', 'Mechanical.2', 'Mechanical.3', 'Microstructure']

for column in columns_to_interpolate:
    df[column] = pd.to_numeric(df[column], errors='coerce')

# Sort the DataFrame based on the 'Heat' column
df.sort_values(by='Heat', inplace=True)

# Interpolate specified columns using spline interpolation
for column in columns_to_interpolate:
    df[column].interpolate(method='spline', order=3, inplace=True)

# Print the modified DataFrame
df
```
